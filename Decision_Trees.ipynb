# Decision Tree


*1. What is a Decision Tree, and how does it work in the context of
classification?*

A Decision Tree is a supervised machine learning algorithm used for solving classification and regression problems. In classification, it predicts a categorical output (e.g., “spam” or “not spam”, “disease” or “no disease”). It works by learning a sequence of simple decision rules from training data and representing them in a tree-like hierarchical structure.

A Decision Tree classifies data using a top-down approach called recursive binary splitting or divide-and-conquer. The main steps include:

Step 1: Select the Best Feature to Split Data
The algorithm selects the feature that best separates classes. Selection is based on impurity measures:

 - Gini Index (CART)
 - Entropy (ID3/C4.5)
 - Information Gain

Step 2: Split the Dataset

Based on the chosen feature, the data is divided into subsets.
Example:
If “Age > 30?” is the best split, two branches will be created:

Yes → samples with Age > 30

No → samples with Age ≤ 30

Step 3: Repeat Recursively

At each new node:

 - Algorithm again checks which feature gives the best split

 - Dataset keeps getting divided into smaller groups

This continues until:

 - Classes become pure (all samples of the same class)

 - No more features remain

 - Tree reaches stopping criteria (depth limit, minimum samples, etc.)

Step 4: Assign Class Labels at Leaf Nodes

Once splitting stops, each leaf node gets a final classification label.
For example, if 90% samples at a leaf are “Spam”, that node predicts Spam.

 - Advantages of Decision Trees

    - Easy to understand and interpret

    - Suitable for both numerical and categorical data

    - No need for data scaling

    - Works well with large datasets

    - Provides clear decision rules
- Disadvantages

    - Prone to overfitting

    - Small changes in dataset may drastically change the tree

    - Biased toward features with many levels
 - Applications

    - Medical diagnosis

    - Spam detection

    - Credit risk analysis

    - Customer segmentation
    

*2.  Explain the concepts of Gini Impurity and Entropy as impurity measures.
How do they impact the splits in a Decision Tree?*

1. Gini Impurity

Gini Impurity measures the probability that a randomly chosen sample would be incorrectly classified if it were labeled according to the class distribution in the node.

Gini=1∑p^2 (c to i=1)

Where:

c = number of classes

𝑝𝑖= probability of class i in the node

Interpretation :

0 → perfectly pure node (only one class)

Higher value → more mixed data

Example :

Suppose a node has:

4 samples of Class A

6 samples of Class B

pA​=0.4,pB​=0.6

Gini=1−(0.42+0.62)=1−(0.16+0.36)=0.48

2. Entropy

Entropy measures the amount of uncertainty or disorder in a node.

Entropy=∑​pi​log2​pi​  (i=1 to c)

Interpretation :

0 → completely pure node

High entropy → higher randomness and mixture

Example

Using the same class distribution:

Entropy≈0.97

Entropy ranges from: 0 to 1

Comparison Between Gini and Entropy

| Feature   | Gini Impurity                              | Entropy                                  |
| --------- | ------------------------------------------ | ---------------------------------------- |
| Formula   | (1 - \sum p_i^2)                           | (-\sum p_i \log_2 p_i)                   |
| Speed     | Faster, simpler                            | Slower (log operations)                  |
| Splitting | Favor larger class separation              | Slightly more sensitive to class changes |
| Used in   | CART (Classification and Regression Trees) | ID3, C4.5 decision trees                 |


Final Summary

 - Gini Impurity and Entropy measure how mixed the classes are at a node.

 - Decision Trees choose splits that reduce impurity the most.

 - Lower impurity = purer nodes = better classification performance.

 - Entropy is based on information theory, while Gini is computationally simpler.

 - Both measures guide the tree to create the most informative splits.

*3. What is the difference between Pre-Pruning and Post-Pruning in Decision
Trees? Give one practical advantage of using each.*

1. Pre-Pruning

Pre-pruning stops the tree from growing too large by restricting or preventing splits during the training phase.

The algorithm checks before splitting a node whether the split will lead to sufficient improvement.
If not, the split is not performed.

Common Pre-Pruning Techniques

 - Minimum number of samples required to split (min_samples_split)

 - Maximum depth allowed (max_depth)

 - Minimum impurity decrease required

 - Minimum number of samples in a leaf (min_samples_leaf)

2. Post-Pruning

Post-pruning allows the tree to grow fully until it becomes large and potentially overfits, and then unnecessary branches or subtrees are removed after training.

It evaluates the contribution of each subtree using validation data or performance metrics.

Common Post-Pruning Methods

 - Reduced Error Pruning

 - Cost-Complexity Pruning (used in CART)

 - Minimum Description Length (MDL)

Key Differences Between Pre-Pruning and Post-Pruning

| Aspect              | Pre-Pruning                       | Post-Pruning                               |
| ------------------- | --------------------------------- | ------------------------------------------ |
| Timing              | Done **during** tree construction | Done **after** tree is fully grown         |
| Tree size           | Prevents large trees              | Cuts down already large trees              |
| Overfitting control | Prevents overfitting early        | Fixes overfitting after it occurs          |
| Computation         | Fast, less expensive              | Comparatively slower                       |
| Decision basis      | Heuristics during training        | Evaluation using validation set or scoring |
| Flexibility         | Less flexible (may underfit)      | More flexible (tends to perform better)    |


Pre-Pruning restricts the tree growth before it becomes too complex.

🔹 Advantage: Fast and efficient for large datasets.

Post-Pruning trims the fully grown tree to remove unhelpful branches.

🔹 Advantage: Reduces overfitting and improves generalization.


*4. What is Information Gain in Decision Trees, and why is it important for
choosing the best split?*

Information Gain (IG) is a key metric used in decision tree algorithms (ID3, C4.5) to decide which feature should be used to split the data at each step.

It measures the reduction in impurity or uncertainty after the dataset is split using a feature.

Information Gain formula:

IG=Entropy(parent)−Entropy(split)

Intuition Behind Information Gain

 - A good split makes child nodes purer (close to a single class).

 - If a feature creates pure child nodes, entropy decreases a lot.

 - This results in a high Information Gain, meaning the feature is very useful.

Why Information Gain is Important for Choosing the Best Split
 - Helps identify the most informative feature
 - Ensures maximum purity at child nodes
 - Reduces tree depth
 - Prevents unnecessary splits
 - Improves model accuracy

Summary

 - Information Gain measures how much a feature reduces uncertainty in the data.

 - It is calculated using the difference between entropy before and after the split.

 - Decision Trees choose the feature with the highest Information Gain at each node.

 - This leads to more accurate, efficient, and pure splitting of the dataset.


*5. What are some common real-world applications of Decision Trees, and
what are their main advantages and limitations?*

1. Common Real-World Applications of Decision Trees

a. Medical Diagnosis

Decision Trees are used in healthcare to classify diseases based on symptoms, test results, or patient history. They help doctors identify conditions such as diabetes, heart disease, or cancer by following a sequence of medical decision rules.

b. Finance and Credit Risk Assessment

Banks and financial institutions use Decision Trees to evaluate loan applicants by analyzing factors like income, credit score, debt history, and employment status. They help classify applicants into low-risk or high-risk categories.

c. Customer Segmentation and Marketing

Businesses apply Decision Trees to segment customers based on purchasing patterns, demographics, and behavior. This helps in targeted marketing, churn prediction, and product recommendation.

d. Fraud Detection

Decision Trees assist in identifying fraudulent transactions by learning patterns from historical transaction data. They help detect anomalies in credit card use, insurance claims, and online payments.

e. Manufacturing and Quality Control

Factories use Decision Trees to identify the root causes of defects in products. Variables such as machine settings, material quality, and environmental factors can be analyzed to improve production efficiency.

f. Human Resource Management

Decision Trees help in employee performance evaluation, attrition prediction, and recruitment decisions by analyzing features like experience, education, and previous performance records.

g. Agriculture

They are used to classify crops, predict yields, and diagnose plant diseases based on soil properties, weather conditions, and plant characteristics.

2. Main Advantages of Decision Trees

a. Easy to Understand and Interpret

Decision Trees are highly intuitive. The decision-making process is visual and follows simple rules, making them understandable even for non-technical stakeholders.

b. No Need for Feature Scaling

Unlike algorithms such as SVM or KNN, Decision Trees do not require normalization or standardization. They can handle both numerical and categorical data directly.

c. Handles Nonlinear and Complex Relationships

Decision Trees can capture nonlinear interactions between features without requiring mathematical transformations.

d. Works Well with Missing Data

Decision Trees have mechanisms like surrogate splits that allow them to handle missing values effectively.

e. Fast Training and Prediction

Decision Trees are computationally efficient and suitable for large datasets. Splitting operations are performed using simple mathematical calculations.

3. Main Limitations of Decision Trees

a. Prone to Overfitting

A major disadvantage is that Decision Trees tend to fit noise in the data, producing very deep and complex trees. Without pruning, they may generalize poorly to new data.

b. High Variance

Small changes in the training dataset can lead to a completely different tree structure. This instability affects model reliability.

c. Bias Toward Features with Many Levels

Features with many distinct values may dominate splits due to their ability to create smaller subsets, even if they are not meaningful.

d. Limited Predictive Power Compared to Ensembles

While Decision Trees are simple, they usually perform worse than ensemble methods like Random Forests or Gradient Boosted Trees, which reduce overfitting and variance.

e. Difficulty in Handling Linearly Separable Data

For problems where a single linear decision boundary is sufficient, simpler models like logistic regression may perform better than a Decision Tree.

*6. Write a Python program to:*

*● Load the Iris Dataset*

*● Train a Decision Tree Classifier using the Gini criterion*

*● Print the model’s accuracy and feature importances*

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train Decision Tree Classifier using Gini criterion
clf = DecisionTreeClassifier(criterion='gini', random_state=42)
clf.fit(X_train, y_train)

# Predict on test set
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Feature importances
feature_importances = clf.feature_importances_

# Print results
print("Model Accuracy:", accuracy)
print("Feature Importances:", feature_importances)
```

*7. Write a Python program to:*

*● Load the Iris Dataset*

*● Train a Decision Tree Classifier with max_depth=3 and compare its accuracy to
a fully-grown tree.*


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Fully grown Decision Tree
full_tree = DecisionTreeClassifier(random_state=42)
full_tree.fit(X_train, y_train)
full_pred = full_tree.predict(X_test)
full_accuracy = accuracy_score(y_test, full_pred)

# Decision Tree with max_depth = 3
limited_tree = DecisionTreeClassifier(max_depth=3, random_state=42)
limited_tree.fit(X_train, y_train)
limited_pred = limited_tree.predict(X_test)
limited_accuracy = accuracy_score(y_test, limited_pred)

# Print results
print("Accuracy of Fully-Grown Tree:", full_accuracy)
print("Accuracy of Tree with max_depth=3:", limited_accuracy)
```

*8. Write a Python program to:*

*● Load the Boston Housing Dataset*

*● Train a Decision Tree Regressor*

*● Print the Mean Squared Error (MSE) and feature importances*



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

data_url = "http://lib.stat.cmu.edu/datasets/boston"

# Load raw dataset
raw_df = pd.read_csv(data_url, sep=r"\s+", skiprows=22, header=None)

# Convert raw data into usable X (features) and y (target)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]

X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.3, random_state=42
)

regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

importances = regressor.feature_importances_

print("Mean Squared Error (MSE):", mse)
print("Feature Importances:", importances)
```

Write a Python program to:

● Load the Iris Dataset

● Tune the Decision Tree’s max_depth and min_samples_split using
GridSearchCV

● Print the best parameters and the resulting model accuracy


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Decision Tree model
dt = DecisionTreeClassifier(random_state=42)

# Parameter grid for tuning
param_grid = {
    'max_depth': [1, 2, 3, 4, 5],
    'min_samples_split': [2, 3, 4, 5]
}

# Grid Search with 5-fold cross-validation
grid = GridSearchCV(dt, param_grid, cv=5)
grid.fit(X_train, y_train)

# Best model
best_model = grid.best_estimator_

# Accuracy of best model
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Print results
print("Best Parameters:", grid.best_params_)
print("Model Accuracy:", accuracy)
```

10.  Imagine you’re working as a data scientist for a healthcare company that

Explain the step-by-step process you would follow to:

1. Handling missing values

-  Characterize missingness: per-feature missing rate, per-row missingness, patterns (e.g., block missing for a subgroup). Visualize with heatmaps / missingness matrices.
- Consider dropping: Drop features with extremely high missingness (e.g., > 70–80%) unless clinically essential.Drop rows only if missingness is rare and dropping does not bias the sample.
- Create missingness indicators: for features where “missing” may be informative, add a binary indicator column (feature_missing = 0/1).
- Choose imputation methods by feature type and missingness pattern:

Numerical features:

 - Simple: median imputation (robust to outliers) or mean if distributions are symmetric.

 - Model-based: K-Nearest Neighbors imputation, Iterative Imputer (e.g., MICE), or regression imputation if relationships are strong.

Categorical features:

 - Impute with a new category like "Missing" or with the most frequent category.

 - Consider model-based categorical imputation (e.g., using a classifier) for MAR cases.
    - Use pipelines: perform imputation inside a preprocessing pipeline to avoid data leakage (fit imputers on training folds only).
    - Evaluate imputation quality: compare distributions before/after imputation, and (if possible) validate by hiding known values and imputing them to measure error.

2. Encoding categorical features

- Distinguish categorical types:

Nominal (no order) vs ordinal (intrinsic order).
- Encoding strategies:

Ordinal features: use OrdinalEncoder or map categories to integers according to order.

Nominal features with few levels: use OneHotEncoder (drop one level to avoid multicollinearity if required).

High-cardinality categorical features: avoid full one-hot encoding. Use alternatives:

 - Target/mean encoding with careful cross-validation and smoothing to avoid leakage (compute encodings within CV folds).

 - Frequency encoding (replace category by its frequency).

 - Hashing encoder if memory is constrained.
    - Maintain interpretability: for healthcare, simpler encodings are often preferable to preserve explainability.
    - Use sklearn’s ColumnTransformer to apply different encoders to different feature subsets inside a pipeline.

3. Feature engineering & scaling

- Feature creation: encode temporal features (age from birthdate), interaction terms if clinically meaningful, binning continuous variables if that aids interpretability.
- Scaling: Decision Trees do not require feature scaling. Scaling is unnecessary for tree-based models but can be kept if later using distance-based models in an ensemble.
- Handle outliers: decision trees are robust to outliers, but extreme values that are data errors should be corrected or removed.

4. Train a Decision Tree model

Define target and train/test split strategy:

 - Use stratified train/test (and stratified CV) to preserve class proportions.

 - Reserve a hold-out test set (e.g., 10–20%) for final evaluation.

 Baseline model: train a default DecisionTreeClassifier (e.g., criterion='gini') to establish baseline performance.

 Cross-validation: use k-fold stratified CV (e.g., k=5 or 10) for reliable performance estimates.

 Address class imbalance if present:

 - Resampling: SMOTE (synthetic oversampling), random oversampling/undersampling (careful with small datasets).

 - Class weights: set class_weight='balanced' in the DecisionTreeClassifier.

Evaluation metrics should emphasize minority class performance (see Section 7).
5.5. Fit the model within a pipeline that contains imputers and encoders to ensure all preprocessing is applied identically in CV.

5. Hyperparameter tuning

 - Important hyperparameters for Decision Tree:

max_depth (controls tree depth)

min_samples_split (min samples to split an internal node)

min_samples_leaf (min samples in a leaf node)

max_features (max features considered at each split)

criterion (gini or entropy)

class_weight (for imbalance)
- Tuning strategy:

Use GridSearchCV for small, focused parameter grids or RandomizedSearchCV for larger spaces.

Use stratified CV and perform tuning inside the pipeline (so imputers/encoders are refit per fold).

Consider nested cross-validation to produce an unbiased estimate of generalization performance when tuning hyperparameters.

 - Use scoring that aligns with business needs (e.g., roc_auc, recall, or a custom scoring function penalizing false negatives more heavily).

- Regularization: prefer smaller trees (limit max_depth, increase min_samples_leaf) to reduce overfitting and improve interpretability.

6. Model evaluation

Choose evaluation metrics appropriate for healthcare:

 - Primary concern is often sensitivity (recall for disease positive) to minimize false negatives (missing sick patients).

 - Also monitor specificity, precision, F1-score, and ROC AUC.

 - For highly imbalanced data, Precision-Recall AUC may be more informative than ROC AUC.
    - Confusion matrix: always report confusion matrix (TP, FP, FN, TN).
    - Calibrate predicted probabilities: Decision Trees often produce poorly calibrated probabilities. Use CalibratedClassifierCV (Platt scaling or isotonic regression) if probabilities are used for risk thresholds.
    - Uncertainty estimation: consider bootstrap or prediction intervals if needed.
    - Explainability: produce feature importances, partial dependence plots, and model-agnostic explanations (SHAP or LIME) to explain individual predictions to clinicians. Decision Trees are naturally interpretable, but complex trees still require tools.
    - Validation on external data: if available, test model on a geographically or temporally distinct dataset to assess generalization and dataset shift.
    - Statistical tests: compare performance vs baseline or other models using appropriate statistical tests
7. Model deployment and monitoring

- Define decision threshold based on operational constraints (e.g., choose a threshold that yields desired sensitivity).
- Integration: expose model via an API or integrate into electronic health record (EHR) workflow with clinician decision support.
- Logging and audit: log model inputs/outputs, predictions, and clinician actions for auditability and for later monitoring.
- Monitor model performance and data drift: track input feature distributions, prediction distributions, and target rates. Retrain or recalibrate when performance degrades.
- Safety and fallback: provide conservative default actions, human-in-loop review for high-risk predictions, and clear disclaimers on model scope and limitations.
- Regulatory and ethical compliance: maintain documentation, bias analysis, and relevant approvals before clinical use.

8. Business value in a real-world healthcare setting

- Early detection and improved outcomes: timely identification of disease can enable earlier interventions, reducing morbidity and mortality.
- Prioritization and triage: automated risk scoring helps prioritize high-risk patients for immediate review or diagnostic testing, optimizing clinician time.
- Resource allocation and cost savings: focusing tests and treatments on likely positive cases reduces unnecessary testing, lowering costs and improving throughput.
- Personalized care: risk stratification enables tailored care pathways and monitoring intensity.
- Operational efficiency: automated pre-screening reduces administrative burden and improves clinic workflow.
- Data-driven decision support: the model can reveal important predictive features that guide clinical investigations, protocol adjustments, or public health responses.
- Business metrics: improved patient outcomes, reduced readmissions, shorter time-to-diagnosis, reduced per-patient cost, and higher clinician productivity.
